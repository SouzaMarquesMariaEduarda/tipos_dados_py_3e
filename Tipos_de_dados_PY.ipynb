```python
#tipos de dados no python - revisão

#int
idade = 17
print("idade:")
print(idade)
print(type(idade))

#float (ponto flutuante) valores com virgula (ponto separador de decimal)
altura = 1.60

#str (string) texto
nome = "Maria Ingrid"
print("nome:")
print(nome)
print(type(nome))

#bool - booleano: o que é um boleano galera (sim, não, verdade)
# (mentira, ligado, desligado) ou seja só há duas opções
briga = False

print("briga:")
print(briga)
print(type(briga))

#list lista (coleção ordenada)
frutas = ["bluberry", "morango", "manga"]

print("fruta:")
print(frutas)
print(type(frutas))

#tuple - dupla, como por exemplo, coordenadas
coordenadas = (52, 640, 24, 325)

print("coordenadas:")
print(coordenadas)
print(type(coordenadas))

#set - conjunto
numeros = {1, 2, 3, 4, 5, 5, 5} # Changed to set literal

print("numeros:")
print(numeros)
print(type(numeros))

#dict - dicionario
pessoas = {"nome": "Nathingrid", "idade":17}

print("pessoas:")
print(pessoas)
print(type(pessoas))

#None Type = quando não tem nada, mas preciso informar um valor
valor = None

print("valor:")
print(valor)
print(type(valor))
```
